```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('201306-citibike-tripdata.csv') 
data.head()
```

```python
tripduration = data.groupby(['birth year'], as_index=False)['tripduration'].mean()
# tripduration = data.groupby('birth year')['tripduration'].mean()
tripduration.head()
```

```python
topstation = data.groupby(['end station id']).size().reset_index(name='counts')
topstation.sort_values('counts', ascending=False).head(10)
```

```python
bottomstation = data.groupby(['end station id']).size().reset_index(name='counts')
bottomstation.sort_values('counts', ascending=True).head(10)
```

```python
gendertype = data['gender']
gendertype.head()
```

```python
male_count = 0
female_count = 0

for type in gendertype:
    if type == 1:
        male_count +=1
    elif type == 2:
        female_count +=1
print(f'male count: {male_count}, female count: {female_count}')
```

```python
peaktimes = data.groupby(['starttime']).size().reset_index(name='counts')
peaktimes.sort_values('counts', ascending=False).head()
```

```python
peaktimes = data.groupby(['starttime']).size().reset_index(name='counts')
peaktimes.sort_values('counts', ascending=False).head()
```

```python
topstation = data.groupby(['start station id']).size().reset_index(name='counts')
topstation.sort_values('counts', ascending=False).head(10)
```

```python
bottomstation = data.groupby(['start station id']).size().reset_index(name='counts')
bottomstation.sort_values('counts', ascending=True).head(10)
```

```python
usertype = data['usertype']
usertype.head()
```

```python
customer_count = 0
subscriber_count = 0

for type in usertype:
    if type == "Customer":
        customer_count +=1
    elif type == "Subscriber":
        subscriber_count +=1
print(customer_count, subscriber_count)
```
